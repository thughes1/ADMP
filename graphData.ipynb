```python
import pandas as pd
import matplotlib.pyplot as plt

def visualise(dfs):
    df1 = dfs[0]
    df2 = dfs[1]
    df3 = dfs[2]
    
    # Set figure size
    fig = plt.figure(figsize=(20, 10))
    
    x1 = (df1['Vehicle Type'])
    y1 = (df1['Mean pm2.5'])
    
    x2 = (df2['Vehicle Type'])
    y2 = (df2['Mean pm2.5'])
    
    x3 = (df3['Vehicle Type'])
    y3 = (df3['Mean pm2.5'])
    
    # Create the first bar plot
    plt.bar(x1, y1, color='blue',label='2013')

    # Create the second bar plot
    plt.bar(x2, y2, color='red', bottom=y1,label='2016')

    # Create the third bar plot
    plt.bar(x3, y3, color='green', bottom=y2,label='2019')
    
    plt.xlabel('Vehicle Type')
    plt.ylabel('Mean pm2.5')
    
    plt.legend()
    plt.show()
    
def preprocess(df):
    # Remove 'PHV - Diesel', 'PHV - Electric','PHV - Petrol' from 2019 csv file'
    return df.drop([10,11,12],axis=0)

def main():
    # Load an ODS file into a Pandas dataframe
    fig2013 = pd.read_csv('figure2013Mean.csv')
    fig2016 = pd.read_csv('figure2016Mean.csv')
    fig2019 = preprocess(pd.read_csv('figure2019Mean.csv'))
    # Append to list
    dfs = [fig2013,fig2016,fig2019]
    # Visualise
    visualise(dfs)
    
    
main()

```
